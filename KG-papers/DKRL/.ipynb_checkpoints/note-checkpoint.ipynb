**Goal function**
$$
L = \sum_{(h,r,t) \in T}\sum_{(h',r',t') \in T'}\max{(0, \gamma + d(h+r,t) - d(h'+r',t'))} \tag{4}
$$

$$
T' = \{(h',r,t)|h' \in E\} \cup \{(h,r,t')|t' \in E\} \cup \{(h,r',t)|r' \in R\}
$$
